```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = plt.imread("/content/Stranger Things.jpg")
plt.imshow(img)
plt.axis('off')
plt.show
```

```python
#Reshaping the numpy array 3D - 2D i.e. an array of 11321344*3 instead of 4096*2764*3
#Before that we need to preprocess the image pixels data, i.e. normalize the pixels' values
```

```python
w,h,d = img.shape
image_array = img.reshape(w*h,d) #you can also write this as image_array = img.reshape(img.shape[0]*img.shape[1], img.shape[2])
image_array = image_array/255
image_array.shape
```

```python
kmeans = KMeans(n_clusters = 3,random_state = 1)
kmeans.fit(image_array)
```

```python
labels = kmeans.predict(image_array)
```

```python
c = kmeans.cluster_centers_
```

```python
def recreate_image(c,labels,w,h,d):
  image = np.zeros((w,h,d))
  label_idx = 0

  for i in range(w):
    for j in range(h):
      image[i][j] = c[labels[label_idx]]
      label_idx +=1
  return(image)
```

```python
plt.figure(1)
plt.axis('off')
plt.imshow(img)
plt.show
plt.figure(2)
plt.axis('off')
plt.imshow(recreate_image(c,labels,w,h,d))
plt.show
```
